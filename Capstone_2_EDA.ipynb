For the Capstone_2 Exploratory data analysis  I have choosen the data of universities in the united states of America.
In the step below I have imported all the required python libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

I have loaded the csv file in the below step and printed the head of the dataset.

```python
# Load the Excel file
df = pd.read_csv('CollegeScoreCard_19-20.csv')
print(df.head())
```

The step below is to see how the data set is shaped.

```python
# General summary statistics
summary_stats = df.describe()
print(summary_stats)
```

Here i have plotted the missing values in the different columns of the data set.

```python
# Count missing values per column
missing_values = df.isnull().sum()

# Top 10 most missing columns
missing_top10 = missing_values.sort_values(ascending=False).head(10)

# Visualization: Top 10 columns with missing values
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_top10.index, y=missing_top10.values)
plt.xticks(rotation=45)
plt.title("Top 10 Columns with Missing Values")
plt.ylabel("Number of Missing Values")
plt.xlabel("Columns")
plt.show()

```

Here I have plotted one of the main element of the university that is the Tution Fees

```python
# Visualizing distribution of numerical data (e.g., tuition costs)
df['Average Net Price Tuition'].dropna().hist(bins=30, edgecolor='black', alpha=0.7)
plt.title("Distribution of Average Net Price Tuition")
plt.xlabel("Tuition Fees")
plt.ylabel("Frequency")
plt.show()

```

Here i have plotted the heatmap of the data set.For that i have formed the correlation  matrix for the numeric columns.

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])


# Correlation matrix for numeric columns
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Matrix")
plt.show()
```
